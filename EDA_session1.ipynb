```python
final_df = pd.merge(df, df_country, on="Country Code", how="left")
final_df.head(5)
```

```python
final_df.Country.value_counts()
```

```python
#if i want to count values sort y thier index i will just add .index
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

**Pie Chart**

```python
#pie chart for top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.2f%%")
```

**Observation:**

Most of the obervation is lying in India, which means Zomato maxi,um records are from India

```python
final_df.columns
```

```python
#let's make a new group of data to see more observation
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

**Observation**



1.   When Rating is between 4.5 to 4.9 --> Excellent
2.   When Rating is between 4.0 to 3.4 --> Very Good
3.   When Rating is between 3.5 to 3.9 --> Good
4.   When Rating is between 3.0 to 3.4 --> Average
5.   When Rating is between 2.5 to 2.9 --> Average
6.   When Rating is between 2.0 to 2.4 --> Poor





```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8, 4)
sns.barplot(x = "Aggregate rating", y = "Rating color", data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y = 'Rating Count', hue='Rating color', data = ratings,  palette=['White', 'Red', 'Orange', 'Green', 'Yellow', 'Green'])
```

```python
sns.countplot(x = "Rating color", data=ratings, palette=['White', 'Red', 'Orange', 'Green', 'Yellow', 'Green'])
```
